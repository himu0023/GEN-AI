```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import random
from skimage import exposure
from skimage.util import random_noise
from skimage import transform
from cv2 import resize
```

```python
img = mpimg.imread("/content/drive/MyDrive/GEN AI/DL/dog.jpg")
```

```python
plt.imshow(img)
```

## Rescaling

```python
img_rescale = resize(img, (450,500))
```

```python
plt.imshow(img_rescale)
```

```python
img_rescale2 = resize(img, (90,100))
```

```python
plt.imshow(img_rescale2)
```

```python
img_rescale3 = resize(img, (45,60))
```

```python
plt.imshow(img_rescale3)
```

## FLIPPING

```python
horizontal_flip = np.fliplr(img)
```

```python
plt.imshow(horizontal_flip)
```

```python
vertical_flip = np.flipud(img)
```

```python
plt.imshow(vertical_flip)
```

## Rotaion

```python
from skimage import transform
trans_img = transform.rotate(img, random.uniform(-40,40))
plt.imshow(trans_img)
```

```python
img_rotate = transform.rotate(img, random.uniform(-56,15))
plt.imshow(img_rotate)
```
